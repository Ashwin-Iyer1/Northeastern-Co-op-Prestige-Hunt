```python
import json

# load the json from companies_with_rank.json
with open('companies_with_rank.json', 'r') as f:
    companies_with_rank = json.load(f)
    
```

```python
from supabase import create_client, Client
# load the .env filefrom dotenv import load_dotenv
import os
from dotenv import load_dotenv
load_dotenv()
# Supabase credentials
SUPABASE_URL = os.getenv("SUPABASE_ENDPOINT")
SUPABASE_KEY = os.getenv("SUPABASE_PUBLISHABLE_KEY")

# Initialize Supabase client
supabase: Client = create_client(SUPABASE_URL, SUPABASE_KEY)
print("Connected to Supabase!")
```

```python
# Update the Supabase client to use service role key
SUPABASE_SERVICE_KEY = os.getenv("SUPABASE_SERVICE_KEY")
supabase: Client = create_client(SUPABASE_URL, SUPABASE_SERVICE_KEY)

# Transform and insert data into Supabase
companies_data = []
for company in companies_with_rank:
    companies_data.append({
        "company_name": company["name"],
        "elo": company["score"]
    })

# Insert data into the companies table
print(f"Inserting {len(companies_data)} companies into Supabase...")
try:
    # if company already exists, update the elo score instead of inserting a new row
    response = supabase.table("companies").insert(companies_data).execute()
    print(f"Successfully inserted {len(companies_data)} companies!")
except Exception as e:
    print(f"Error: {e}")
```
